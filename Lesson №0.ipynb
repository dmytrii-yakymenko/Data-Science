# Lesson №0 / introductory lesson

###  Розкриття основних понять

**A/B тестування (або спліт-тестування)** — це метод, що використовується для порівняння двох версій чогось (наприклад, веб-сторінки, реклами, продукту) з метою визначення, яка з них є більш ефективною в досягненні певної мети (наприклад, збільшення конверсії, кліків або залученості користувачів).

**Як це працює:**
- **Створення варіантів:** Створюються два варіанти (A і B). Варіант A — це контрольна версія (оригінал), а варіант B — змінена версія (експериментальна).

- **Випадковий розподіл:** Користувачі або учасники випадковим чином розподіляються на дві групи: одна взаємодіє з варіантом A, а інша — з варіантом B.

- **Збір даних:** Протягом тесту збираються дані про поведінку користувачів, такі як кількість кліків, покупки, час перебування на сайті тощо.

- **Аналіз результатів:** Після завершення тесту дані аналізуються, щоб визначити, яка версія була більш успішною в досягненні поставленої мети. Зазвичай використовуються статистичні методи для оцінки значущості результатів.

- **Вибір переможця:** На основі отриманих даних приймається рішення про те, чи варто впроваджувати зміну з варіанту B, чи залишити варіант A.

**Переваги A/B тестування:**
- **Дані для прийняття рішень:** Допомагає базувати рішення на реальних даних, а не на припущеннях.
- **Оптимізація:** Дає змогу покращити продукти чи сервіси шляхом поступового тестування та вдосконалення.
- **Зниження ризиків:** Зменшує ризик впровадження змін, які можуть негативно вплинути на користувацький досвід.

**Способи реалізації:**
1) **Компіляція** - одразу запускається весь код:
    - У цьому випадку весь код програми перетворюється на машинний код або байт-код перед виконанням. Компілятор аналізує весь код, оптимізує його і генерує виконуваний файл.
    - Переваги:
        - Швидке виконання програми, оскільки код вже скомпільований.
        - Більш оптимізоване використання ресурсів.
    - Недоліки:
        - Довший час підготовки (компіляція), перш ніж програма може бути запущена.
        - Виникнення помилок на етапі компіляції, що може вимагати повторної компіляції.
2) **Інтерпретація** - запускає код по частинам:
    - Інтерпретатор виконує код по частинам, зчитуючи та виконуючи його рядок за рядком під час виконання програми.
    - Переваги:
        - Швидкий цикл розробки, оскільки зміни в коді можна тестувати негайно без повторної компіляції.
        - Легкість виявлення помилок, оскільки вони можуть бути помічені під час виконання.
    - Недоліки:
        - Зазвичай повільніше виконання програми, оскільки код обробляється під час виконання.
        - Вищі вимоги до ресурсів, оскільки код потрібно інтерпретувати в реальному часі.

**Парадигма програмування** у контексті Data Science парадигма програмування може включати кілька ключових аспектів:

1. **Ітеративний підхід:** Data Science часто вимагає експериментування та ітерацій, що означає, що ви будете повторювати етапи аналізу даних, моделювання і валідації. Цей підхід нагадує гнучкі методології розробки програмного забезпечення.

2. **Об'єктно-орієнтоване програмування (ООП):** Використання класів і об'єктів для структурування кодів, що дозволяє створювати модульні та повторно використовувані компоненти, особливо в бібліотеках для машинного навчання (наприклад, scikit-learn у Python).

3. **Функціональне програмування:** Багато інструментів для обробки даних, таких як pandas у Python, використовують функціональні парадигми, такі як застосування функцій до колекцій даних (наприклад, map, filter, reduce).

4. **Декларативний підхід:** У таких мовах, як SQL, ви описуєте, що хочете отримати з даних, а не як це робити. Це робить запити до баз даних зрозумілішими та менш схильними до помилок.

5. **Паралельне та розподілене програмування:** Для роботи з великими даними часто використовують паралельні та розподілені обчислення (наприклад, Apache Spark), що дозволяє обробляти великі обсяги даних ефективно.

**IDE (інтегроване середовище розробки)** для Data Science допомагає програмістам писати, тестувати та відлагоджувати код, а також працювати з даними. Ось кілька популярних IDE та інструментів, які часто використовуються в Data Science:

- **Jupyter Notebook** - Інтерактивне середовище, яке дозволяє створювати документи, що містять код, візуалізації та текстові пояснення. Це особливо корисно для аналізу даних та візуалізації.

- **Sublime Text** — це легкий, але потужний текстовий редактор, який часто використовується програмістами для написання коду, в тому числі в Data Science. 

- **PyCharm** - Потужне IDE для Python, яке має багато функцій, зокрема підтримку для аналізу даних, відлагодження та інтеграцію з бібліотеками для машинного навчання.

### Пояснення спеціально-технічних термінів:

**Репозиторій** — це місце для зберігання, організації та управління кодом, даними або іншими файлами, зазвичай у контексті програмування та розробки програмного забезпечення. Репозиторії часто використовуються в системах контролю версій, таких як Git.

**Markdown** — це легка мова розмітки, яка дозволяє форматувати текст у простому текстовому файлі. Вона була створена Джоном Грубером у 2004 році з метою спростити написання форматованих документів, які легко читаються в звичайному текстовому вигляді. Markdown часто використовується для написання документації, README-файлів, блогів, а також в системах управління контентом.

**LMS (Learning Management System)** — це система управління навчанням, яка використовується для планування, виконання та оцінювання навчальних процесів.

**Pet project** — це невеликий, зазвичай особистий проект, який реалізується для задоволення, експериментів або навчання. Ці проекти часто не мають комерційної мети, а скоріше слугують способом розвивати свої навички або реалізувати цікаві ідеї.
